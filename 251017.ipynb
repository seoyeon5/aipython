```python
#3-2
students = ["철수", "영희", "민수"]
scores = [85, 92, 78]
grades = ["B", "A", "C"]

print("\n학생 성적표:")
for name, score, grade in zip(students, scores, grades):
 print(f"{name}: {score}점 ({grade}등급)")
```
