```python
#조선 왕 27명의 묘호가 담긴 리스트를 사용하여, 순번을 입력하면 해당 왕의 묘호
#묘호를 입력하면 순번을 알려주는 양방향 사전 프로그램을 만듭니다.
#사용자의 입력이 숫자인지 문자열인지를 판별하여 두 기능을 하나의 프로그램으로 통합

dynasties = ['태조','정조','태종','세종','문종','단종','세조','예종','성종','연산군','중종','인종','명종','선조','광해군','인조','효종','현종','숙종','경종','영조','정조','순조','헌종','철종','고종','순종']

ans = input('순번 또는 왕의 이름을 입력')

#순번
if ans.isdigit():
   i=int(ans)
   if 1 <= i <= len(dynasties):
    print(dynasties[i-1])
   else:
     print('잘못입력함')
#이름
else:
  if ans in dynasties:
    print(dynasties.index(ans)+1)
  else:
      print('없음')
```

```python
#총점, 평균, 가장 높은 점수 학생이름

score ={'홍길동':100, '이순신':90, '강감찬':80}

#총점
sum(score.values())
```

```python
#평균
sum(score.values())/len(score)
```

```python
#최고점
max(score.values())

for i in score:
  if score[i] == max(score.values()):
    print(i)
```
