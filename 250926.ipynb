```python
num = int(input("숫자를 입력하세요: "))
if num % 2 == 0:
  print("짝수입니다") # 이 문장은 위의 조건이 '참'일 때만 실행됩니다.
else:
  print("홀수입니다")
```

```python
height1 = int(input("첫 번째 학생의 키를 입력하세요: "))
height2 = int(input("두 번째 학생의 키를 입력하세요: "))

# 만약 height1이 height2보다 크다면
if height1 > height2:
  diff = height1 - height2
# 그렇지 않다면 (height2가 더 크거나 같다면)
else:
  diff = height2 - height1
print("두 학생의 키 차이는", diff, "입니다.")
```

```python
score =int(input("점수를 입력하세요: "))

if score>=90:
  print("A")
elif score>=70:
  print("B")
elif score>=60:
  print("C")
elif score<60:
  print("D")
```

```python
sum = 0 # 합계를 저장할 변수를 0으로 초기화

for a in range(1, 101):
  sum = sum + a # 기존 sum 값에 현재 a 값을 더해서 다시 저장
print("1부터 100까지의 합은", sum, "입니다.")
```

```python
sum1 = 0 #짝수합
sum2 = 0 #홀수합

for a in range(1, 101):
  if a%2==0:
    sum1 = sum1 + a
  else:
    sum2 = sum2 + a
print("1부터 100까지의 짝수합은", sum1, "입니다.")
print("1부터 100까지의 홀수합은", sum2, "입니다.")
```

```python
#숫자를 입력받아서 총합을 구하는 프로그램을 작성, 단 0이 입력되면 종료되고 최종 총합을 출력

sum = 0

while True:
  num = int(input("숫자를 입력하세요"))
  sum = sum + num
  if num==0:
    break

print("총합은",sum,"입니다.")
```
